```python
import pandas as pd
import numpy as np
import re

# Example DataFrame structure with flight data
df = pd.DataFrame({
    'departure_time': ['12:10', '12:10', '09:50', '17:00', '19:20', '10:10', '23:40', '23:55', '23:50', '17:25',
                       '17:25', '11:10', '14:20', '14:20', '21:05', '13:35', '19:55', '19:35', '16:50', '21:05',
                       '17:10', '12:55', '23:50'],
    'arrival_time': ['12:55', '14:55', '12:55', '14:30', '16:50', '07:40', '21:05', '21:25', '21:35', '20:25',
                     '22:50', '19:40', '13:53', '13:53', '22:35', '15:35', '22:35', '22:35', '19:10', '15:10+1',
                     '16:30+1', '12:30', '07:36+1'],
    'total_time': ['15 hours 45 minutes', '17 hours 45 minutes', '18 hours 5 minutes', '12 hours 30 minutes', '12 hours 30 minutes', 
                   '12 hours 30 minutes', '12 hours 25 minutes', '12 hours 30 minutes', '12 hours 45 minutes', '18 hours 0 minutes', 
                   '20 hours 25 minutes', '23 hours 30 minutes', '14 hours 33 minutes', '14 hours 33 minutes', '16 hours 30 minutes', 
                   '17 hours 0 minutes', '17 hours 40 minutes', '18 hours 0 minutes', '17 hours 20 minutes', '33 hours 5 minutes', 
                   '38 hours 20 minutes', '14 hours 35 minutes', '22 hours 46 minutes'],
    'price': [42593, 42593, 42593, 57621, 63021, 63021, 56861, 63021, 57621, 49393,
              47393, 50197, 65301, 65955, 56204, 61716, 56204, 56204, 61216, 53202, 
              51025, 74031, 57043]
})

# Function to convert flight duration from string to minutes
def convert_to_minutes(time_str):
    hours, minutes = map(int, re.findall(r'\d+', time_str))
    return hours * 60 + minutes

# Apply the conversion function to the total_time column
df['total_time_minutes'] = df['total_time'].apply(convert_to_minutes)

# Initialize parameters for the Ant Colony Optimization algorithm
num_ants = 10  # Number of ants
num_iterations = 100  # Number of iterations
evaporation_rate = 0.5  # Pheromone evaporation rate
alpha = 1  # Importance of pheromone
beta = 2  # Importance of heuristic

# Initialize pheromone matrix
pheromone = np.ones(len(df))

# Heuristic function: inverse of the flight duration
heuristic = 1 / df['total_time_minutes']

# Variables to track the best time-cost and best price routes
best_time_cost = float('inf')
best_route_time_cost = None
best_price = float('inf')
best_route_price = None

# Ant Colony Optimization algorithm
for iteration in range(num_iterations):
    all_routes = []
    all_time_costs = []
    all_prices = []
    
    for ant in range(num_ants):
        # Select a route based on pheromone and heuristic values
        probabilities = (pheromone ** alpha) * (heuristic ** beta)
        probabilities = probabilities / probabilities.sum()
        route = np.random.choice(len(df), p=probabilities)
        all_routes.append(route)
        
        # Calculate the combined cost (flight duration * price)
        time_cost = df.iloc[route]['total_time_minutes'] * df.iloc[route]['price']
        all_time_costs.append(time_cost)
        
        # Record the price
        price = df.iloc[route]['price']
        all_prices.append(price)
    
    # Update the pheromone matrix
    pheromone = (1 - evaporation_rate) * pheromone
    
    for i, route in enumerate(all_routes):
        pheromone[route] += 1 / all_time_costs[i]
    
    # Select the route with the lowest combined cost
    current_best_time_cost = min(all_time_costs)
    if current_best_time_cost < best_time_cost:
        best_time_cost = current_best_time_cost
        best_route_time_cost = all_routes[np.argmin(all_time_costs)]
    
    # Select the route with the lowest price
    current_best_price = min(all_prices)
    if current_best_price < best_price:
        best_price = current_best_price
        best_route_price = all_routes[np.argmin(all_prices)]
    
    print(f'Iteration {iteration + 1}, Best Time Cost: {best_time_cost}, Best Price: {best_price}')

# Output the best routes and costs
print('Best Route (lowest combined cost):', best_route_time_cost)
print('Best Time Cost (lowest combined cost):', best_time_cost)
print('Best Route (lowest price):', best_route_price)
print('Best Price (lowest price):', best_price)
```
